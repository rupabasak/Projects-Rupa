```python
data = df[['Close']]
data
```

```python
data.reset_index(inplace=True)
```

```python
from prophet import Prophet
model_pr = Prophet()
```

```python
model_pr.fit(data)
```

```python
fig = model_pr.plot(forecast)
plt.title('Reliance Stock Price Prediction for Next 30 Days using Prophet')
plt.xlabel('Date')
plt.ylabel('Price')
plt.show()
```

```python
# Cross Validation
from prophet.diagnostics import cross_validation
```

```python
df_cv = cross_validation(model_pr, initial='730 days', period='180 days', horizon='365 days')
```

```python
# 1. model: The Prophet model that has been fitted to the data.

# 2. initial: The size of the initial training period. 
#             This is the duration from the start of the time series data to the first cutoff point.

# 3. period: The spacing between cutoff dates.
#            This defines how often the model will be retrained and tested on a new subset of data.

# 4. horizon: The forecast horizon. 
#             This is the duration for which the forecast is made from each cutoff point.
```

```python
# conducting dickey fuller test to check stationary.
```

```python
from statsmodels.tsa.stattools import adfuller 
```

```python
test_result = adfuller(df['Close']) 
```

```python
# ho: The data is not stationary.
# ha: The data is stationary.
```

```python
def adfullet_test(Close):
    result = adfuller(Close)
    labels = ['ADF Test', 'P-value', 'Lags Used', 'Number Of Observation']
    for value, label in zip(result, labels):
        print(label + ': ' + str(value))
    if result[1] <= 0.05:
        print("Null Hypothesis is rejected. The data is Stationary")
    else:
        print("Null Hypothesis is accepted. The data is not Stationary")
```

```python
adfullet_test(df['Close'])
```

```python
# The assumption is rejected. (0.99 !<= 0.05)
```

# Sarimax

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
import warnings
warnings.filterwarnings('ignore')
```

```python
data2 = df[['Close']]
data2
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
data2['Close'] = scale.fit_transform(data2[['Close']])
```

```python
model = SARIMAX(data2, order=(1,1,1), seasonal_order=(1,1,1,12))
model
```

```python
model_fit = model.fit(disp=False) 
print(model_fit.summary()) 
```

```python
forecast_steps = 30
forecast = model_fit.get_forecast(steps=forecast_steps)
forecast_index = pd.date_range(start=data2.index[-1], periods=forecast_steps + 1, freq='B')[1:]
forecast_values = scale.inverse_transform(forecast.predicted_mean.values.reshape(-1, 1))
```

```python
forecast_df = pd.DataFrame(data=forecast_values, index=forecast_index, columns=['Forecast'])
forecast_df
```

```python
plt.figure(figsize=(14,7))
sns.scatterplot(forecast_df, s=100)
plt.xlabel('Date')
plt.ylabel('Forecast')
plt.title('Forecast Data')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.plot(forecast_df, color='orange', linestyle='--', marker='*')
plt.xlabel('Date')
plt.ylabel('Forecast')
plt.title('Forecast Data')
plt.show()
```

```python
plt.figure(figsize=(14, 7))
plt.plot(data2.index, scale.inverse_transform(data2), label='Actual')
plt.plot(forecast_df.index, forecast_df['Forecast'], label='Forecast')
plt.title('Reliance Stock Price Prediction for Next 30 Days using SARIMAX')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```

```python
# Calculate RMSE
```

```python
from sklearn.metrics import mean_squared_error
actual = data2['Close'][1:]
predicted = model_fit.fittedvalues[1:]

rmse = np.sqrt(mean_squared_error(actual, predicted))
print(f"Root Mean Squared Error (RMSE): {rmse:.4f}")
```

```python
# Calculate MSE
```

```python
from sklearn.metrics import mean_squared_error
actual = data2['Close'][1:]
predicted = model_fit.fittedvalues[1:]

mse = mean_squared_error(actual, predicted)
print(f"Root Mean Squared Error (RMSE): {mse:.4f}")
```

```python
# Calculate MAE
```

```python
from sklearn.metrics import mean_absolute_error
actual = data2['Close'][1:]  
predicted = model_fit.fittedvalues[1:]

mae = mean_absolute_error(actual, predicted)
print(f"Mean Absolute Error (MAE): {mae:.4f}")
```

```python
# Calculate MAPE
```

```python
actual = data2['Close'][1:]  
predicted = model_fit.fittedvalues[1:]
mape = np.mean(np.abs((actual - predicted) / actual)) * 100
print(f"Mean Absolute Percentage Error (MAPE): {mape:.2f}%")
```
